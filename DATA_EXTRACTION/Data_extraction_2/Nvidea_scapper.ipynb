```python
def get_text_from_webpage(url):
    try:
        # Send a GET request to the URL
        response = requests.get(url)

        # Check if the request was successful
        response.raise_for_status()

        # Parse the HTML content using BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find all the text elements within the HTML
        text_elements = soup.find_all(text=True)

        # Filter out empty and whitespace strings, join the rest into a single string
        text_data = ' '.join(text.strip() for text in text_elements if text.strip())

        return text_data

    except requests.exceptions.RequestException as e:
        print(f"Error: {e}")
        return None
```

```python
from bs4 import BeautifulSoup
from bs4.element import Comment
import urllib.request


def tag_visible(element):
    if element.parent.name in ['style', 'script', 'head', 'title', 'meta', '[document]']:
        return False
    if isinstance(element, Comment):
        return False
    return True


def text_from_html(body):
    soup = BeautifulSoup(body, 'html.parser')
    texts = soup.findAll(text=True)
    visible_texts = filter(tag_visible, texts)  
    return u" ".join(t.strip() for t in visible_texts)


url = "https://docs.omniverse.nvidia.com/nucleus/latest/index.html"
url = "https://docs.omniverse.nvidia.com/kit/docs/kit-manual/latest/guide/kit_architecture.html"

html = urllib.request.urlopen(url).read()
print(text_from_html(html))
```

```python
filterd_urls = []
clear_url_tags = [] 
kit_content = ""

for tail in href_list:
    new_url = base_url + tail
    
        
    try:
        html = urllib.request.urlopen(new_url).read()
        soup = BeautifulSoup(html, 'html.parser')
            
        temp = tail.split("_")
        
        new_t = []
        for t in temp:
            if t == "kit":
                continue
                
            elif t.endswith(".html"):
                new_t.append(t[:-5])
                
            else:
                new_t.append(t)
                        
        new_tail = ""
        for t in new_t:
            new_tail += t
            new_tail += "-"
            
        new_tail = new_tail[:-1]

        
        ele = soup.findAll("section", {"id": new_tail})
        
        if len(ele) == 0:
            filterd_urls.append([new_url, new_tail])
            
        else:
            for tag in ele:
                kit_content += tag.text.strip()
    #             kit_content += "\n\n"
                kit_content += "-*-"*100

            clear_url_tags.append(new_tail)
        
    except:
        continue

```

```python
def get_content(url, id_text):
    
    try:
        html = urllib.request.urlopen(url).read()
        
    except:
        return None

    soup = BeautifulSoup(html, 'html.parser')


    ele = soup.findAll("section", {"id": id_text})

    if len(ele) == 0:
        return None

    content = ""
    for tag in ele:
        content += tag.text.strip()
    
#     print(content)
#     print("\n\n")
    return content
    
```

```python
import requests
from bs4 import BeautifulSoup

def get_text_in_div_with_id(url, div_id):
    try:
        # Send a GET request to the URL
        response = requests.get(url)

        # Check if the request was successful
        response.raise_for_status()

        # Parse the HTML content using BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find the <div> element with the specified id
        div_element = soup.find('div', {'id': div_id})

        if div_element:
            # Get all the text within the <div> element
            text_data = div_element.get_text()
            return text_data.strip()

    except requests.exceptions.RequestException as e:
        print(f"Error: {e}")
        return None

# Example usage
if __name__ == "__main__":
    url = "https://docs.omniverse.nvidia.com/dev-guide/latest/get-started.html"  # Replace with the URL of the webpage you want to scrape
    div_id = "get-started"
    text_in_div = get_text_in_div_with_id(url, div_id)
    if text_in_div:
        print(text_in_div)
        print("-*-"*100)
```
