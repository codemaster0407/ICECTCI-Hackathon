````python
def get_completion(prompt, model="gpt-3.5-turbo"):
    messages = [{"role": "user", "content": prompt}]
    response = openai.ChatCompletion.create(
        model=model,
        messages=messages,
        temperature=0, 
    )
    return response.choices[0].message["content"]
````

````python
import traceback
qa = []
except_index = []
for i in range(len(temp)):
    
    try:
        text = temp['data'][i]
#         print(text)
        template_string = """Translate the text that is delimited by triple backticks into a format that is into question and answers about NVIDIA documentations. \
                            text: ```{text}```. Generate upto ten to twenty reasonable questions for training a chatbot per text. Don't include personal view questions and generate question on a more generic view.Give a list of questions and answers.Format the output as a list of JSON with the following keys: \
                            question
                            answer
                            """
        prompt_template = ChatPromptTemplate.from_template(template_string)
        customer_messages = prompt_template.format_messages(text = text)
        customer_response = chat(customer_messages)
        qa.append(customer_response.content)
        
        print(f" {i} is done")
    except Exception as e:
        traceback.print_tb(e.__traceback__)
        except_index.append(i)
        print(i)
        continue

    
````
