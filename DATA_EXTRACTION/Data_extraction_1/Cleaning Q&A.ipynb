```python
import re
def find_exceptions(text):
    text = str(text)
    matches = re.findall(r'"(.*?)"', text)
    for match in matches:
#         print(match)
        if match == 'Q1':
            return True
    return False
```

```python
import json
```

```python
def json_data_clean(text):
    try:
        data = json.loads(text)

        # Separate questions and answers into separate lists
        questions = [item['question'] for item in data]
        answers = [item['answer'] for item in data]

        return questions,answers
    except:
        return None, None

#     print("Questions:", questions)
#     print("Answers:", answers)
```

```python
questions, answers = [],[]
exceptions = []
for index in range(len(temp['q&a'])):
    text = temp['q&a'][index]
    q,a = json_data_clean(text)
    if q==None and a == None:
        exception.append(index)
    questions.append(q)
    answers.append(a)
    
```

```python
def flatten_list(lst):
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(flatten_list(item))
        else:
            result.append(item)
    return result
```
